**Remove all % of electric cars with a value of 0**

```python
df = df[df['% of electric cars'] != 0]
df
```

#Questions

1.   What is the correlation between the charging ports and the percentage of electric cars?
2.   If I increase the number of charging ports, will the number of electric cars increase?


3.   How is the population of an area linked to the number of charging ports?





```python
# A chart highlighting No of Charging ports against % of electric cars
# sample any random 20
random_sample_df = df.sample(n=80)
random_sample_df.plot(kind='scatter', x='% of electric cars', y='No of Charging ports', legend=False)
```

## Hypothesis
Number of charging ports are high when the percentage of electric cars is high.

Make another chart for population against % of electric cars

```python
random_pop_sample_df = df.sample(n=80)
random_pop_sample_df.plot(kind='scatter', x='% of electric cars', y='Population', legend=False)
```
